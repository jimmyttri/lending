```python
y_final.sum()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression(penalty = 'l2', solver = 'sag', C = 100, max_iter = 20)
```

```python
log.fit(X_val, y_val)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
preds = log.predict(X_final)
```

```python
print(confusion_matrix(y_final, preds))

print(classification_report(y_final, preds))
```

```python
plt.plot(log.coef_.T, 'o', label="C=1")
```
